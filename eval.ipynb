## 🎯 Objective
This notebook evaluates the performance of the grant and tender matcher system.

The system:
- Parses multilingual tender documents (EN/FR)
- Matches them to business profiles
- Ranks top-5 relevant tenders
- Generates short summaries in the user's preferred language

---

## 📊 Evaluation Metrics

We use:

### 🔹 MRR@5 (Mean Reciprocal Rank)
Measures how high the first correct match appears in the ranking.  
Higher is better (max = 1.0)

### 🔹 Recall@5
Measures how many expert-selected tenders appear in the top 5.  
Higher is better (max = 1.0)

---

## ⚙️ Approach Summary

The matcher uses:
- TF-IDF + BM25 combined scoring
- Translation for cross-lingual matching (FR ↔ EN)
- Lightweight rule-based boosts:
  - Sector match
  - Language match
  - Region match
- CPU-only design (fast and deployable)
